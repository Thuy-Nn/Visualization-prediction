# Modelling

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# Split the data into features and target variable

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
class Net1(nn.Module):
    def __init__(self, input_size, output_size):
        super(Net1, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(input_size, 32),
            nn.ReLU(),
            nn.Linear(32, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, output_size)
        )        

    def forward(self, x):
        return self.model(x)

#multi-layer perceptron (MLP) model- rewrite MLPClassifier
class Net2(nn.Module):
    def __init__(self, input_size, output_size):
        super(Net2, self).__init__()
        self.activation = nn.ReLU()

        layers = []
        layer_sizes = [input_size] + list((100, 50)) + [output_size]

        # Create fully connected layers with specified hidden sizes
        for i in range(len(layer_sizes) - 2):
            layers.append(nn.Linear(layer_sizes[i], layer_sizes[i+1]))
            layers.append(self.activation)

        # Output layer (without activation if we are using CrossEntropyLoss)
        layers.append(nn.Linear(layer_sizes[-2], layer_sizes[-1]))

        self.model = nn.Sequential(*layers)

    def forward(self, x):
        return self.model(x)
```

```python
configurations = [{
        'model': Net1,
        'criterion': nn.CrossEntropyLoss,
        'optimizer': optim.Adam,
        'lr': 0.01,
        'bs': 32
    }, {
        'model': Net2,
        'criterion': nn.CrossEntropyLoss,
        'optimizer': optim.SGD,
        'lr': 0.001,
        'bs': 32
    }
]
```

```python
inputs = torch.tensor(X_train.values, dtype=torch.float)
print(inputs.shape)

# Convert the target Series to a tensor with long data type
labels = torch.tensor(y_train.values, dtype=torch.long)

encoded_labels = []
for i in range(y_train.shape[0]):
    arr = [0.0] * 11
    arr[labels[i]] = 1.0
    encoded_labels.append(arr)    

# print(labels[:10])
# print(encoded_labels[:10])

encoded_labels = torch.tensor(encoded_labels)
```

```python
from torch.utils.data import DataLoader, TensorDataset
import matplotlib.pyplot as plt

for conf in configurations:
    net = conf['model'](6, 11)
    criterion = conf['criterion']()
    optimizer = conf['optimizer'](net.parameters(), lr=conf['lr'])
    print(f'===== Model={net.__class__.__name__}, Criterion={criterion.__class__.__name__}, Optimizer={optimizer.__class__.__name__}, lr={conf["lr"]}, bs={conf["bs"]}')

    dataset = TensorDataset(inputs, encoded_labels)
    train_dataloader = DataLoader(dataset, batch_size=conf['bs'], shuffle=True)
    
    print('Start training ...')

    # Loop over dataset multiple times. Pytorch accumulates gradients per default, so it is required to
    # zero the parameter gradients before each iteration with 'zero_grad()' function.
    conf['loss_vals'] = []
    for epoch in range(100):
        losses = []
        for batch_inputs, batch_labels in train_dataloader:
            optimizer.zero_grad()
            outputs = net(batch_inputs)
            loss = criterion(outputs, batch_labels)
            loss.backward()
            optimizer.step()
            losses.append(loss.item())
        epoch_loss = np.average(losses)
        conf['loss_vals'].append(epoch_loss)
        if (epoch + 1) % 10 == 0:
            print(f'Epoch {epoch + 1}, Loss: {epoch_loss:.4f}')

    # do the inference
    print('Start testing ...')
    net.eval()

    X_test_tensor = torch.tensor(X_test.values, dtype=torch.float)
    y_test_tensor = torch.tensor(y_test.values, dtype=torch.long)
    
    with torch.no_grad():
        
        outputs = net(X_test_tensor)
        predictions = torch.argmax(outputs, dim=1)
    print(f'Accuracy: {torch.sum(predictions == y_test_tensor)/predictions.shape[0]:.4f}')

# plot the losses
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
ax1.plot(range(len(configurations[0]['loss_vals'])), configurations[0]['loss_vals'])
ax2.plot(range(len(configurations[1]['loss_vals'])), configurations[1]['loss_vals'])
ax1.set_title(f'{configurations[0]['model'].__name__} - Training Loss')
ax2.set_title(f'{configurations[1]['model'].__name__} - Training Loss')
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Loss")
ax2.set_xlabel("Epoch")
ax2.set_ylabel("Loss")
plt.show()
```
